```python
#le profit pour chaque magasin en soustrayant les dépenses marketing du revenu
```

```python
startups['Profit'] = startups['Revenue'] - startups['Marketing Spend']
startups
```

```python
#Calcul du retour sur les dépenses marketing (ROMS) et conversion en valeur décimale pour analyser l'efficacité des campagnes marketing
```

```python
startups['ROMS'] = round((startups['Profit'] / startups['Marketing Spend']) * 100, 2)
```

```python
startups['ROMS%'] = startups['ROMS'] / 100
```
